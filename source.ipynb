## MATRIX FUNCTIONS

```python
class Matrices:#class of matrix methods
    
    ###################################################
    #### BASIC MATRIX ARITHMATIC ###################
    ##############################################
    
    def add(matrixA, matrixB):
        #Adds two matrices together, component by component
        #RESTRICTIONS:
        #Can only add two matrices of equal size
        newArray = []
        for rowIndex in range(len(matrixA)):
            #goes through each row and creates a new row filled with the sums of each index
            newRow = []
            for colIndex in range(len(matrixA[rowIndex])):
                newRow.append(matrixA[rowIndex][colIndex] + matrixB[rowIndex][colIndex])
            newArray.append(newRow)
        return newArray

    def multScalar(matrix, scalar):
        #Multiplies a single vector (meaning a single column) by a scalar
        #RESTRICTIONS:
        #Input matrix can only have one column, or else returned matrix will only have the first row * scalar
        newMatrix = []
        for rows in matrix:
            newMatrix.append([rows[0]*scalar])
        return newMatrix
    
    def multMatrix(matrixA, matrixB):
        #Multiplies two matrixes together using linear combinations (returns A*B)
        #If a (height x width) c x n and an n x r matrix are combined, the resulting matrix will be size c x r
        #RESTRICTIONS:
        #Matrix A must have as many *columns* (width) as Matrix B has *rows* (height)
        newMatrix = []
        
        for columnB in range(len(matrixB[0])):
            #Goes by b-columns
            newCol = []
            for rowA in range(len(matrixA)):
                newValue = 0
                for rowB in range(len(matrixB)):
                    newValue += matrixA[rowA][rowB] * matrixB[rowB][columnB]
                    #multiplying the column values of B by all the row values of A
                newCol.append(newValue)
                #adding them to the column array
            newMatrix.append(newCol)
        ##resulting matrix is backward because we added the sum of the numbers to a row
        return Matrices.transpose(newMatrix)
        
    
    ###################################################
    #### SINGLE MATRIX OPERATIONS ##############
    ##############################################
    
    def transpose(matrix):
        #Switches rows and columns of any matrix
        #(works with any 2d array too lol)
        #If a matrix is n x m, transpose (matrix) returns an m x n matrix
        newMatrix = []
        for col in range(len(matrix[0])):
            newRow = []
            for row in range(len(matrix)):
                newRow.append(matrix[row][col])
            newMatrix.append(newRow)
        return newMatrix
    
    def determinant(matrix):
        #Finds the mathematical determinant of a n x n (square) matrix
        #Takes in a matrix, returns a scalar
        #RESTRICTIONS:
        #Must be a square matrix
        if len(matrix) == len(matrix[0]) == 2:#Can return ad-bc
            return (matrix[0][0] * matrix[1][1] - matrix[1][0] * matrix[0][1])
        else:#Returns the sum of the array of the value * (-1)^row * det(otherPart)
            terms = []
            for i, row in enumerate(matrix):#i: row #, row: row entry @ intersection
                multiplier = (-1)**(i)
                value = row[0]
                smaller = [rowVals[1:] for index, rowVals in enumerate(matrix) if index != i]
                terms.append(multiplier * value * Matrices.determinant(smaller))
            return sum(terms)
        
    def inverse(matrix):
        #Finds the inverse of a n x n reducable matrix
        #Takes a matrix, returns a matrix
        #RESTRICTIONS:
        #Must be perfectly reducable and square
        for row, rowArr in enumerate(matrix):
            pass
    
    def rowReduce(matrix):
        #Puts the matrix in row echelon form
        newMatrix = matrix.copy()
        size = min([len(newMatrix), len(newMatrix[0])])#Get the smaller of the two, to get the size of the "box"
        
        diagonalized = True
        for diag in range(size):#makes sure each diagonal element is filled
            diagonalized = False if newMatrix[diag][diag] == 0 else diagonalized
        
        if diagonalized == False:#arrange into an easily reducable matrix
            order = []
            for row in range(size):
                for col in range(size):
                    if matrix[row][col] != 0:
                        order.insert(col, row)
                        break
            print(order)
            for i, rowNumber in enumerate(order):
                newMatrix[i] = matrix[rowNumber]
        print(matrix)
        print(newMatrix)
        for col in range(size):
            for row in range(size):
                if row != col and newMatrix[row][col] != 0 and 0 != newMatrix[col][col]:
                    coeff = -newMatrix[row][col] / newMatrix[col][col] #coefficients to make everything but the target variable 0 when added
                    newMatrix[row] = Matrices.rowAdd(newMatrix[col], newMatrix[row], coeff)
        return newMatrix    
    #####################################################
    #### REDUCTION OPERATIONS ###########################
    ##############################################
    def rowAdd(row1, row2, constant = 1):
        #Returns lambda*A*B for each element
        return [constant*row1[i] + row2[i] for i in range(len(row1))]
```

## String-interpretation functions

```python
class Read:
    
    def matrix(inputStr):
        ##Converts a string notating a matrix into a usable matrix (in the form of a 2d array)
        #################
        ##Example:
        #
        #Input:
        #Read.matrix("[[30, 4],[-3, 5]]")
        #
        #Output:
        #[[30, 4],[-3, 5]]
        string = "".join([letter for letter in inputStr if letter != " "])
        indexBrackets = [[i for i in range(len(string)) if string[i] == "["],[i for i in range(len(string)) if string[i] == "]"]]
        insides = []
        for i in range(1, len(indexBrackets[0])):
            #Looks in between bracket pairs
            insides.append(string[1+indexBrackets[0][i] : indexBrackets[1][i-1]])
        ##Grabs the strings between the bracket pairs and turns them into lists of floats
        return [[float(number) for number in string[1+indexBrackets[0][i]:indexBrackets[1][i-1]].split(",")] for i in range(1, len(indexBrackets[0]))]
        
    def unpeel(inputStr, beginChr = "(", endChr = ")"):
        ##Unpeels an outer layer of brackets from a bunch of nested brackets, returning an array of inner strings
        #String should inculde the first opening of brackets
        #IF THERE ARE MULTIPLE INNER STRINGS, ALL ARE RETURNED
        ################
        #Example:
        #
        #Input:
        #Read.unpeel("det([[30, 4],[-3, 5]]) + mult_s(2, [[30, 4],[-3, 5]])")
        #
        #Output:
        #['[[30, 4],[-3, 5]]', '2, [[30, 4],[-3, 5]]']
        layer = 0
        
        array = []
        index = 0
        for letter in inputStr:
            #iterates through the input string, only adds numbers if they are at least in 1 layer of parantheses
            #creates a new element in output if a new parantheses is opened
            if letter == endChr:
                layer -=1
                if layer == 0:#moves the index indicator forward when the parentheses is closed
                    index += 1 #tracks layer
            
            if layer > 0:#checks layer, adds letter to the string if its deep enough
                array[index] += letter
                
            
            if letter == beginChr:
                if layer == 0:#when parentheses is opened, instiantiates a new string because new entry is inevitable
                    array.append("")
                layer += 1 #tracks layer
                
        return array#Returns an array of all strings within outer-layer brackets
    
    def splitOutside(inputStr, splitter = ",", beginChr = "(", endChr = ")"):
        #Splits a string without harming the insides, notated by any bracket system
        #Returns a list of what was between the "splitters"
        #RESTRICTIONS:
        #Assumes only one pair of parentheses
        #Eliminates spaces
        ################
        #Example:
        #
        #Input:
        #Read.splitOutside("mult_s(2 + det(A), B + C) + D", "+")
        #
        #Output:
        #["mult_s(2+det(A),B+C)", "D"]
        word = "".join([letter for letter in inputStr if letter != " "])
        layer = 0
        innerStrings = [""]
        index = 0
        for i in range(len(word)):
            if word[i] == endChr:
                layer -=1

            if word[i] == splitter and layer == 0:
                index +=1
                innerStrings.append("")
            else:
                innerStrings[index] += word[i]


            if word[i] == beginChr:
                layer += 1 

        return innerStrings
    
    def checkOutside(word, char, beginChr = "(", endChr = ")"):
        #Boolean expression testing whether there is a certain character on outside of the bracket system
        #Returns a True expression if "char" is found, else returns false
        ##########
        #Example:
        #
        #input:
        #Read.checkOutside("hello, call me (724-300-3000)", "-")
        #
        #output:
        #False
        found = False
        layer = 0
        for i in range(len(word)):
            if word[i] == endChr:
                layer -=1
            if word[i] == char and layer == 0:
                found = True


            if word[i] == beginChr:
                layer += 1 
        return found
```
